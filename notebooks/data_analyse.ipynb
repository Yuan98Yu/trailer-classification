```python
from torch.utils.data import DataLoader
import torchvision.transforms as tt
from torchvision.datasets import ImageFolder

from data import show_batch

data_dir = './data'
```

```python
stats = ((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))

train_transform1 = tt.Compose([
    tt.Resize([400, 400]),
    # tt.RandomHorizontalFlip(),
    # tt.RandomVerticalFlip(),
    # tt.RandomRotation(90),
    # tt.ColorJitter(brightness=0.5, contrast=0.5, saturation=0.5, hue=0.5),
    tt.ToTensor(),
    tt.Normalize(*stats)
])
```

```python
full_dataset = ImageFolder(data_dir)
```

```python
dl = DataLoader(full_dataset, batch_size=64)

full_dataset.transform = train_transform1
show_batch(dl)
```

```python
min_x, min_y = 999, 999
for image, label in full_dataset:
    min_x, min_y = min(image.size[0], min_x), min(image.size[1], min_y)
print(min_x, min_y)
    
```
